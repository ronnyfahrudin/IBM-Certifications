# 2. Adding Lattitude and Longitude

```python
import pandas as pd
import numpy as np
import geocoder
```

```python
df = pd.read_csv('toronto.csv')
df
```

```python
df.describe()
```

```python
def latlong(postal_code):
    lati_long_coords = None
    while(lati_long_coords is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
        lati_long_coords = g.latlng
    return lati_long_coords
    
latlong('M1B')[0]
```

```python
df
```

```python
lat= []
long = []
for i in df['Postal Code'].values:
    lat.append(latlong(i)[0])
    long.append(latlong(i)[1])
```

```python
df['Latitude'] = lat
df['Longitude'] = long
df
```

```python
df.to_csv('toronto_latlong.csv', index = False)
```
